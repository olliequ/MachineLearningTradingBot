### 1-B: Dealing with Missing Values [0.75 marks]

You will write <b>two functions</b> to replace missing values using the strategies below based on the different attribute types (categorical, continuous):

   * <b>Continuous features</b>: For each feature find the <b>average feature value</b> in the train dataset 
   * <b>Categorical features</b>: For each feature find the <b>most frequent value</b> in the train dataset 

and a <b>third function</b> that replaces the missing values in the train and test datasets with the most frequent (categorical features) and average values (continuous features) that you found in the train dataset.
 

**Your tasks**

You will implement three functions
1. `num_mean_feat`
<ul>
  <li>input:  instances with only continuous features</li>
  <li>output: a list containing the average value for each continuous feature (aka substitute values)</li>
</ul> 
    
2. `cat_most_feat`
<ul>
    <li>input: instances with only categorical features</li>
    <li>output: a list containing the most frequent feature value for each categorical feature (aka substitute values)</li>
</ul>

3. `transform` 
<ul>
    <li>input:
        <ul>
            <li>a list of instances</li>   
            <li>replacement list containing the substitute values for each feature</li>
        </ul>
    </li>
    <li>output: the transformed instances</li>
</ul>

<b> You should implement these functions from scratch yourself. You may use native Python libraries such as math, collections, or numpy to help you, but you may not use existing implementations.</b>

````python
import copy
from collections import Counter
import numpy as np
from numpy.core.fromnumeric import shape, size
from sklearn import preprocessing

def num_mean_fit(X):
    replacement = []
    np_array = np.array(X)
    no_of_instances = len(np_array)
    no_of_features = len(np_array[0])
    for i in range(0, no_of_features):            # Number of features.
        features_added = 0
        total_nonmissing_values = 0
        for j in range(0, no_of_instances):       # Number of instances.
            try:
                float(np_array[j][i])
            except ValueError:
                continue
            features_added += float(np_array[j][i])
            total_nonmissing_values += 1
        average_feature = features_added/total_nonmissing_values
        replacement.append(average_feature)
    return replacement

def cat_most_fit(X):
    replacement = []
    np_array = np.array(X)
    no_of_instances = len(np_array)
    no_of_features = len(np_array[0])
    for i in range(0, no_of_features):            # Number of features.
        category_values = []
        for j in range(0, no_of_instances):       # Number of instances.
            category_values.append(np_array[j][i])
        occurence_count = Counter(category_values)
        winning_category = occurence_count.most_common(1)[0][0]
        replacement.append(winning_category)
    return replacement

def transform(X, replacement):
    trans_X  = copy.deepcopy(X)
    np_array = np.array(X)
    no_of_instances_to_correct = len(X)
    no_of_features_to_check = len(np_array[0])
    for i in range(0, no_of_features_to_check):              # Number of features.
        for j in range(0, no_of_instances_to_correct):       # Number of instances.
            if np_array[j][i] == '?':
                replacement_value = replacement[i]
                trans_X[j][i] = replacement_value
    return trans_X
````

### Question 3-A: Implement a helper function [0.75 mark]

**Your tasks**

You will implement a function that calculates the error rate for each attribute if we select that attribute as the one rule for classification.
1. `feature_error`
<ul>
  <li>input:
      <ul>
          <li> features</li>
          <li> labels</li>
      </ul>
  </li>
  <li>output:
      <ul>
          <li> a dictionary of the form {feature: error rate on the train} where each 
        feature is denoted by its ID (position in the train dataset)</li>
          <li> a dictionary of the form {feature value: majority class label} where the majority class label is either "obese" or "notObese" and feature values are the best feature's possible values (feature with smallest error rate)</li>
      </ul>
    </li>   
</ul> 

<b> You should implement this function from scratch yourself. You may use native Python libraries like math, numpy, collections to help you, but you may not use existing implementations.</b>


````python
# Takes in a dictionary for one features value. Returns, for that value, which label was seen the most, the total count of that label, 
# and how many times that value is present in that feature column.
def dict_values_total(dictionary):
    total = 0
    for value in dictionary.values():
        total += value
    all_values = dictionary.values()
    max_value = max(all_values)
    max_key = max(dictionary, key=dictionary.get)
    return max_key, max_value, total

# Main function.
def feature_error(features, labels):
    feat_error = {}
    feature_dictionaries = {}  
    feat_rule = {}
    label_options = list(set(labels))
    label_options_dict = dict.fromkeys(label_options, 0)
    for i in range(0, len(features[0])):
        # Need to make a dictionary for each value in this feature. 
        value_dictionaries = {}
        for j in range(0, len(features)):
            value = features[j][i]
            if value not in value_dictionaries:
                a = copy.deepcopy(label_options_dict)
                value_dictionaries[value] = a
            label = labels[j]
            value_dictionaries[value][label] += 1
        feature_dictionaries[i] = value_dictionaries
        total_count = 0
        total_correct = 0
        for key in value_dictionaries:
            max_key, max_value, total = dict_values_total(value_dictionaries.get(key))
            total_count += total
            total_correct += max_value

        total_wrong = total_count - total_correct
        feature_total_errors = total_wrong/total_count
        feat_error[i] = feature_total_errors

    best_feature = min(feat_error, key=feat_error.get)
    best_value_dictionaries = feature_dictionaries.get(best_feature)
    for key in best_value_dictionaries:
        dictionary = best_value_dictionaries.get(key)
        most_frequent_label = max(dictionary, key=dictionary.get)
        feat_rule[key] = most_frequent_label
    return feat_error, feat_rule
````
